```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


def plot_average_reward_2 (filename_1, filename_2,filename_3,filename_4,   
                           color1, color2, color3, color4   ):
    plt.figure(figsize=(20,10))
    step = []
    reward = []
    with open(filename_1,"r") as f:
        line = f.readlines()
    for i in line:
        l = i.split()
        if l==[]:
            break
        if(int(l[1]) % 100000==0   ):
            step.append(int(l[1])  // 100000  )
        reward.append(float(l[2]))
    label = "DQN"
 
    avg = [sum(reward[i*10:(i+1)*10])/10 for i in range(len(reward)//10)] #每十个计算均值
    plt.plot(step, avg, color1, label =label  )
  
    
    step = []
    reward = []
    with open(filename_2,"r") as f:
        line = f.readlines()
    for i in line:
        l = i.split()
        if l==[]:
            break
        if(int(l[1]) % 100000==0   ):
            step.append(int(l[1])  // 100000  )
        reward.append(float(l[2]))
        
    label ="DDQN"
    
    avg = [sum(reward[i*10:(i+1)*10])/10 for i in range(len(reward)//10)]
 
    plt.plot(step, avg, color2, label =  label  )
    
    step = []
    reward = []
    with open(filename_3,"r") as f:
        line = f.readlines()
    for i in line:
        l = i.split()
        if l==[]:
            break
        if(int(l[1]) % 100000==0   ):
            step.append(int(l[1])  // 100000  )
        reward.append(float(l[2]))
       
    label = "Averaged DQN(N=2)"
    
    avg = [sum(reward[i*10:(i+1)*10])/10 for i in range(len(reward)//10)]
     
    plt.plot(step, avg, color3, label =  label  )
    
    step = []
    reward = []
    with open(filename_4,"r") as f:
        line = f.readlines()
    for i in line:
        l = i.split()
        if l==[]:
            break
        if(int(l[1]) % 100000==0   ):
            step.append(int(l[1])  // 100000  )
        reward.append(float(l[2]))
         
    label = "Maxmin DQN(N=2)"
    
    avg = [sum(reward[i*10:(i+1)*10])/10 for i in range(len(reward)//10)]
 
    plt.plot(step, avg, color4, label =  label  )
    
 
    
    plt.ylabel("Average Rewards", fontsize=20)
    plt.xlabel("Training Steps(10^5)", fontsize=20)
    plt.ylim(0,20)
    # {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
    plt.legend(fontsize="large")

 
    plt.savefig(  'result.png')
    
```
